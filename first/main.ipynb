## Задача 1

```python
students = [
    {'name': 'Алексей', 'grade': 85, 'city': 'Москва'},
    {'name': 'Мария', 'grade': 92, 'city': 'Казань'},
    {'name': 'Иван', 'grade': 78, 'city': 'Москва'},
    {'name': 'Елена', 'grade': 88, 'city': 'Иннополис'},
    {'name': 'Дмитрий', 'grade': 55, 'city': 'Москва'},
]
```

```python
goodStudents = list(filter(lambda student: student['grade'] > 80, students))

print(goodStudents)
```

```python
print(list(map(lambda goodStudent: goodStudent['name'].upper(), goodStudents)))
```

```python
from functools import reduce

print(reduce(lambda x, y: x + y, map(lambda student: student['grade'], students)) / len(students))
```

## Задача 2

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

data = {
'product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
'price': [100, 150, 100, 200, 150, 100, 200, 150],
'quantity': np.random.randint(1, 10, size=8)
}

df = pd.DataFrame(data)
print(df)
```

```python
df['revenue'] = list(map(lambda row: row['price'] * row['quantity'], df.to_dict('records')))

df.head()
```

```python
df.groupby('product')['revenue'].sum()
```

```python
mean_qty = df['quantity'].mean()

filtered_df = df[df.apply(lambda row: row['quantity'] > mean_qty, axis=1)]

print(f"Продажи с quantity > среднего ({mean_qty}):")
print(filtered_df)
```

## Задача 3

```python
def get_list(string):
    return list(map(int, string.split(' ')))

print(get_list('1 2 2 3 5 3 1 3'))
```

```python
def sort_func(get_list_func, string):
    array = get_list_func(string)
    array.sort()

    return array

print(sort_func(get_list, '3 1 4 1 5 9 2 6'))
```

## Задача 4

```python
import time
from functools import wraps

def timer(function):
    @wraps(function)

    def wrapper(*args, **kwargs):
        start = time.time()

        result = function(*args, **kwargs)

        end = time.time()

        finishTime = end - start

        print(f"Функция {function.__name__} выполнилась за {finishTime:.4f} сек.")

        return result

    return wrapper
```

```python
@timer
def calculate_sum():
    summa = 0

    for i in range(1, 1000001):
        summa += i

    return summa
```

```python
result = calculate_sum()

print(result)
```
